```python
def categorize_business(row):
    if row['Type'] in ['Pubs', 'Coffee & Tea', 'Bar & Grills', 'Bars', 'Cafes', 'Dive Bars', 'Breweries', 'Diners', 'Juice Bars & Smoothies',
                       'Sports Bars', 'Bakeries', 'Cat Cafes',  'Bakery & Cafes', 'Delis', 'Taverns', 'Dance Clubs', 'Cat cafes', 'Taverns']:
        return 'Social'
    elif row['Type'] in ['Libraries', 'Community Centers', 'Churches', 'Religious Organizations', 'Parks', 'Dog Parks']:
        return 'Community'
    elif row['Type'] in ['Record Stores', 'Bookstores', 'Thrift Stores', 'Shopping Centers', 'Chocolatiers & Shops', 'Bar & Record Stores']:
        return 'Retail'
    elif row['Type'] in ['Recreation Centers', 'Skating Rinks', 'Skate Parks']:
        return 'Recreation'
    else:
        return 'Unknown'
```

```python
attribute['Category'] = attribute.apply(categorize_business, axis=1)
attribute
```

```python
income_third= current_income.merge(third_count, on='postal_code', how='inner')
income_third
```

```python
bins = [0, 55000, 65000, 75000, 85000, 95000, 105000, 115000, 155000]  # Define bins for home values
labels = ['< 55k' ,'55k - 65k', '65k - 75k', '75k - 85k', '85k - 95k', '95k - 105k','105k - 115k', '115k <']  # Bin labels
income_third['income_value_bin'] = pd.cut(income_third['Income'], bins=bins, labels=labels)
```

```python
income_third
```

```python
income_distribution = income_third.groupby('income_value_bin')['count'].sum().reset_index()
income_distribution
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(income_distribution['income_value_bin'], income_distribution['count'], color='skyblue')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom')
plt.title('Total Number of Third Places by Income Range')
plt.xlabel('Income Range')
plt.ylabel('Total Number of Third Places')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()
```

```python
from scipy.stats import f_oneway
```

```python
anova_result = f_oneway(
    income_third[income_third['income_value_bin'] == '< 55k']['count'],
    income_third[income_third['income_value_bin'] == '55k - 65k']['count'],
    income_third[income_third['income_value_bin'] == '65k - 75k']['count'],
    income_third[income_third['income_value_bin'] == '75k - 85k']['count'],
    income_third[income_third['income_value_bin'] == '85k - 95k']['count'],
    income_third[income_third['income_value_bin'] == '95k - 105k']['count'],
    income_third[income_third['income_value_bin'] == '105k - 115k']['count'],
    income_third[income_third['income_value_bin'] == '115k - 125k']['count'],
    income_third[income_third['income_value_bin'] == '125k - 135k']['count'],
    income_third[income_third['income_value_bin'] == '135k - 145k']['count'],
    income_third[income_third['income_value_bin'] == '145k - 155k']['count']
)
```

```python
print(f'ANOVA Test Result: F-statistic={anova_result.statistic:.2f}, p-value={anova_result.pvalue:.4f}')
```
